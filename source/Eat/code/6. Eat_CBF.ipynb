```python
import pandas as pd
import numpy as np
import math
```

```python
# get size of vector
def mag(x): return math.sqrt(sum(i**2 for i in x))
```

```python
res_matrix = pd.read_csv('../data/res_matrix.csv')
user_matrix = pd.read_csv('../data/user_matrix.csv')
group_count = pd.read_csv('../data/group_count.csv',names=['group','count'])

res_matrix.set_index('p_id',drop=True,inplace=True)
user_matrix.rename(columns = {"Unnamed: 0": "u_id"}, inplace=True)
user_matrix.set_index('u_id', drop=True, inplace=True)
group_count.set_index('group', drop=True, inplace=True)
```

```python
p_m = pd.DataFrame(np.array(res_matrix) * (1/np.array(group_count['count'])), index=res_matrix.index)
p_m = p_m.iloc[:,:].div(p_m.sum(axis=1), axis=0).fillna(0)
u_m = pd.DataFrame(np.array(user_matrix) * (1/np.array(group_count['count'])), index=user_matrix.index)
u_m = u_m.iloc[:,:].div(u_m.sum(axis=1), axis=0).fillna(0)
```

```python
user_id = int(input('user_id : '))
```

```python
u_vec = u_m.loc[user_id,:]
u_size = mag(u_vec)
u_size
```

```python
cos_sim = []

u_vec = u_m.loc[user_id,:]
u_size = mag(u_vec)

for i in range(len(p_m)):
    p_vec = p_m.iloc[i,:]
    p_size = mag(p_vec)
    cos_sim.append(np.dot(p_vec, u_vec)/(p_size*u_size))
```

```python
p_m['cbf_score'] = cos_sim
p_m
```

```python
pd.DataFrame(p_m['cbf_score']).sort_values(by=['cbf_score'], axis=0, ascending=False).head(100)
```

```python
pd.DataFrame(p_m['cbf_score']).sort_values(by=['cbf_score'], axis=0, ascending=False).to_csv('../data/predictions/'+str(user_id)+'_cbf.csv')
```
