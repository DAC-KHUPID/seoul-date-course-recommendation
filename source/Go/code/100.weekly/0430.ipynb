```python
import pandas as pd
from gensim import models
```

```python
df = pd.read_csv('./kakao_reviews_each_place.csv')
df
```

```python
import fasttext.util
# fasttext.util.download_model('en', if_exists='ignore')
ft = fasttext.load_model('cc.en.300.bin')
```

```python
review_all=''
for i in df['review_en']:
    review_all = str(review_all)+ '. '+ str(i);
```

```python
review_all
```

```python
import nltk
token = nltk.word_tokenize(review_all)
```

```python
review_all= review_all.replace(",","")
review_all= review_all.replace(".","")
review_all= review_all.replace("!","")
review_all= review_all.replace("~","")
review_all= review_all.replace("\\","")
pos_token = nltk.pos_tag(review_all.split())
```

```python
pos_token
```

```python
#size: vector-size
#window: ì•ë’¤ë¡œ ëª‡ê°œì˜ ë‹¨ì–´ê¹Œì§€ ì°¸ê³ í•´ì„œ ë²¡í„°ë¥¼ ì¶”ì¶œí• ì§€
#min_count: ë‹¨ì–´ ë¹ˆë„ìˆ˜ì— ë”°ë¼ ê²°ê³¼ count x
#worekrs: cpu core
#sg: sigmoid or CBow
ft_model = models.FastText(token, size=300, window=3, min_count=5, workers=4, sg=1)
```

```python
token_ls = token.copy()
```

```python
#ì¤‘ì²©ë˜ëŠ” ê²ƒ ì§€ìš°ê¸°
token_ls= list(set(token_ls))

print(len(token)) #84820
print(len(token_ls)) #5217
```

```python
stop_words = ['Yeonnam-dong', 'êº…', '^^', '/', 'u', 'ì™€', 'ì ›', 'ë¼ë©´', 'ë•ã…‹', 'Ha', 'Oh']
```

```python
new_df = pd.DataFrame() 
for i in pos_token:
    if (i[1] == 'JJ') & (str(i[0]) not in stop_words):
        f_vec = ft_model.wv.word_vec(i[0])
        new_df[i[0]] = f_vec
```

```python
#must be close
print('--must be close--')
print(ft_model.wv.similarity('close', 'far'))
print(ft_model.wv.similarity('delicious', 'tasty'))
print(ft_model.wv.similarity('atmosphere', 'mood'))

#!
print('--!--')
print(ft_model.wv.similarity('silent', 'quiet'))
print(ft_model.wv.similarity('loud', 'quiet'))

#must be far
print('--must be far--')
print(ft_model.wv.similarity('speaker', 'treasure'))
print(ft_model.wv.similarity('delicious', 'computer'))
```

```python
import numpy as np
x = new_df.T
```

```python
from sklearn.cluster import KMeans
cluster_count=20
model = KMeans(n_clusters=cluster_count)
model.fit(x)
```

```python
model.predict(x)
Y=x.copy()
Y['kmeans_id'] = model.predict(x)
```

```python
km = list(Y['kmeans_id'])
for i in range (0, cluster_count):
    print(i,': ', km.count(i))
```
