```python
import pandas as pd
import nltk
```

```python
place_df = pd.read_csv('../data/all_places_reviews.csv')
word_cluster = pd.read_csv('../data/clustered.csv')
```

```python
del word_cluster['Unnamed: 0']
```

```python
place_df.dropna(inplace=True)
```

```python
del place_df['review_kr']
```

```python
place_df.reset_index(inplace=True)
```

```python
word_cluster
```

```python
cluster_count = 100;
for i in range (0,len(place_df)):
    print(i)
    p_id = place_df['p_id'][i]
    review = place_df['review_en'][i]
    token_ls = nltk.word_tokenize(str(review));
    total_count = 0
    res_df['p_id'][i] = p_id
    for j in range (0,cluster_count):
        count = len(set(token_ls).intersection(list(word_cluster[str(j)])))
        total_count = total_count + count
        res_df[j][i] = count
    res_df['total_count'][i] = total_count
```

```python
res_df = pd.DataFrame(index=range(0,len(place_df)), columns=range(0, 100))
```

```python
res_df['p_id'] = 0
res_df['total_count'] = 0
```

```python
place_df
```

```python
res_df.to_csv('../data/place_matrix.csv', index=False)
```
