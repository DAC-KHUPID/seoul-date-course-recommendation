```python
import pandas as pd
```

# NCF data

```python
file=pd.read_csv("../data/movie_watcha_review.csv",index_col=0)
file.head()
```

```python
data={
    'u_id':file['user'],
    'p_id':file['movie_id'],
    'rating':file['rate']
}
score_board=pd.DataFrame(data, columns=['u_id','p_id','rating'])
score_board.sort_values(by=['u_id'], axis=0,inplace=True)
score_board.reset_index(drop = True, inplace=True)
score_board
```

```python
score_board.to_csv('../data/score_board.csv', index=False)
```
